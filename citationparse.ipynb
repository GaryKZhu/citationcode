```python
import nltk
import sqlite3
import os
import re
import sys
from nltk import sent_tokenize

#declare globals
citations = []
sentences = []
sentenceid = []
totcount = 0 
filename = ""
arxiv_id = ""

def getSentences(conn):
    #query database to get sentences 
    
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM sentences where arxiv_id = " + "\"" + arxiv_id + "\"")
    row = cursor.fetchone()

    while row is not None:
        sentences.append(row[1])
        citations.append(row[2])
        sentenceid.append(row[3])
        row = cursor.fetchone()

    cursor.close()


def findCitationPattern(conn):
    
    #this is to find the FTFTF pattern. 
    
    counter = 0
    prev = -1
    cur = -1
    
    
    for i in range(0, len(sentences)):
        #update variables
        if(citations[i] == 1):
            prev = cur
            cur = i
            counter += 1
        
        #find ftftf  pattern
        elif(citations[i] == 0 and cur == i-1 and prev == i-3 and counter >= 2 and i >= 4 and citations[prev-1] == 0): #very messy...
            print(sentences[prev-1])
            create_citation(conn, (arxiv_id, sentences[prev-1], sentences[prev], sentences[prev+1], sentences[i-1], sentences[i], sentenceid[prev], sentenceid[cur])) 

    
def create_citation(conn, citation):
    
    #function to append the citation pattern into sqlite
    
    sql = ''' INSERT INTO citations  (arxiv_id,
       first_sentence,
       second_sentence,
       third_sentence,
       fourth_sentence,
       fifth_sentence,
       second_sentence_id,
       fourth_sentence_id)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?) '''
    cur = conn.cursor()
    cur.execute(sql, citation)
    conn.commit()
    return

if(__name__ == "__main__"):
    
    #read input
    directory = "/Users/brian/projects/rnn/thesis-master/unarXive-2020/papers"
    filecounter = 0
    sconn =  sqlite3.connect("sentence.db")
    cconn = sqlite3.connect("citation.db")
    
    for filename in os.listdir(directory):
        
        arxiv_id = filename[:len(filename)-4]
        print(arxiv_id)
        #read sentences from db
        getSentences(sconn)

        #find all patterns of FTFTF
        findCitationPattern(cconn)

        #add the number of sentences to the total number of sentences
        totcount += len(sentences)
        filecounter += 1
        print(filecounter)
        sentences.clear()
        citations.clear()
        sentenceid.clear()        
        
    sconn.close()
    cconn.close()
```
